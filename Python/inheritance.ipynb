* Inheritance is a feature of OOP (Object-Oriented Programming) where one class can use the properties and methods of another class.

* Parent Class / Base Class → The class whose properties are inherited.
* Child Class / Derived Class → The class that inherits properties.

        Inheritance helps in:

* Code reusability
* Reducing duplicate code
* Easy maintenance


Types of Inheritance in Python


| Type                     | Meaning                                    |
| ------------------------ | ------------------------------------------ |
| Single Inheritance       | One parent → One child                     |
| Multiple Inheritance     | Multiple parents → One child               |
| Multilevel Inheritance   | Grandparent → Parent → Child               |
| Hierarchical Inheritance | One parent → Multiple children             |
| Hybrid Inheritance       | Combination of different inheritance types |


```python
# Single Inheritance 

class Collage:
    pass 
class Student(Collage):
    pass
```

```python
class Collage:
    collage_name = "ARYA MAIN CAMPUS"
    collage_address = "jaipur"
    collagesp = 42
    def show(self):
        print(self.collage_name)
        print(self.collage_address)
        print(self.collagesp)
        
class Student(Collage):
    student_name = "Rohit"
    student_rollnumber = 74114
    student_number = 74141520525
    student_branch = "CSE"
st = Student() 
st.show()
print(st.student_branch)
```

```python
# Multilevel Inheritance
class Person:
    pass 
class Teacher(Person):
    pass 
class HOD(Teacher):
    pass

```

```python
class Person:
    def person_info(self):
        print("i am person")
class Teacher(Person):
    def teacher_info(self):
        print(" i am a teacher")
class HOD(Teacher):
    def hod_info(self):
        print(" i am hod ")
h =HOD()
print(h.hod_info())
print(h.person_info())
print(h.teacher_info() )       
```

```python
# Multiple Inheritance 
class Developer:
    pass   

class Manager:
    pass    
class Teamleader(Developer, Manager):
    pass
 
```

```python
# Multiple Inheritance 
class Developer:
    def codding(self):
        print("bahya ye code likhega ")
        

class Manager:
    def meeting(self):
        print("eska ek hi kam hai vo hai meeting esko kuch aata hai nhi ")    
class Teamleader(Developer, Manager):
    def project(self):
        print("project ko headle krna ")
        
team =Teamleader()

team.codding()
```
