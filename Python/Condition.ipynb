### Condition Statements

**Condition Statements** are used to make decisions in a program.
They execute different blocks of code based on *conditions*.

**<span style="color:red">Types of Condition Statements in Python</span>**

              

* if statement

* if-else statement

* if-elif-else statement

* Nested if statement
