### Looping in Python
* A loop in Python is used to repeat a block of code multiple times until a condition is satisfied.

        Types of Loops in Python

        Python mainly has 2 types of loops:

1. for loop

2. while loop


            1. for loop
* Used when the number of iterations is known.

```python
lst = [1,2,5,8]
for y in  lst:
    print(y)### 
```

```python
for x in range(1,20):
    print(x)
```

```python
for x in range(2,22,2): ### 2 is starting point and 22 ending point or 2 is gap 
    print(x)
```

```python
tpl = (1,2,3,2,5,2,5,0,5,0,5,5,5)
for item in tpl:
    print(item)
```

```python
lst = [1,2,3,5]
for x in lst:
    if x == 3:
        break
    print(x)
```

```python
lst = [1,2,3,5]
for x in lst:
    if x == 3:
        continue
    print(x)
```

```python
x = 1 
while x<=5:
    print(x)
    x+=1
```
